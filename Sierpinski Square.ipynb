```python
import matplotlib.pyplot as plt
import math
```

```python
def nextPoint (x, y, angle):
    a = math.pi * angle / 180
    x2 = (int) (round(x + (1 * math.cos(a))));
    y2 = (int) (round(y + (1 * math.sin(a))));
    return x2, y2
```

```python
def expand(axiom, rule, level):
    for l in range(0,level):
        a2 = ""
        for c in axiom:
            if c in rules:
                a2 += rules[c]
            else:
                a2 += c
        print(f"level {l} length {len(axiom)}")
        axiom = a2
    return axiom
```

```python
axiom = "F+XF+F+XF"
rules = { "X" : "XF-F+F-XF+F+XF-F+F-X"}
angle = 90

# dragon curve
axiom = "FX"
rules = {
    "X" : "X+YF+",
    "Y" : "-FX-Y"
        }
angle = 90

xp = [1]
yp = [1]
```

```python
for c in expand(axiom, rules, 15):
    if c == "F":
        xn, yn = nextPoint(xp[-1], yp[-1], angle)
        xp.append(xn)
        yp.append(yn)
    elif c == "-":
        angle = angle - 90
        if angle < 0:
            angle = 360 + angle
    elif c == "+":
        angle = (angle + 90) % 360

plt.plot(xp, yp)
```
