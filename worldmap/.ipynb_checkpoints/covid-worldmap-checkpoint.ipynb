```python
import csv
import pygal
from pygal.style import Style
import pandas as pd
from IPython.display import SVG, HTML, display
from country_codes import get_country_code
```

```python
filename = "../data/covid-death-confirmed.csv"
df = pd.read_csv(filename)
df
```

```python
def get_index(v):
    if v < 5:
        return 0
    if v < 10: 
        return 1
    if v < 20:
        return 2
    
    return 3
```

```python
cc_population = [dict() for x in range(4)]

for index, row in df.iterrows():
    code = get_country_code(row['country'])
    if code:
        p = row['percent_deaths']
        cc_population[get_index(p)][code] = p
    else:
        print(f"no code for /{row['country']}/ found")
```

```python
custom_style = Style(
    font_family='googlefont:Raleway',
    tooltip_font_size=8
)
```

```python
wm = pygal.maps.world.World(height=300, style=custom_style)
labels = [".lt.  5%", ".lt. 10%", ".lt. 20%", "more than 20%"]

wm.title = "World Deaths of Confirmed Cases, by Country"
for idx in range(0, len(labels)):
    if cc_population[idx]:
        wm.add(labels[idx], cc_population[idx])
    
wm.render_to_file('covid.svg')
```
