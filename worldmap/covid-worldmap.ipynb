```python
import csv
import pygal
import pandas as pd
from IPython.display import SVG, HTML, display
from country_codes import get_country_code
```

```python
filename = "../data/covid-death-confirmed.csv"
df = pd.read_csv(filename)
df
```

```python
def get_index(v):
    if v < 5:
        return 0
    if v < 10: 
        return 1
    if v < 20:
        return 2
    
    return 3
```

```python
cc_population = [dict() for x in range(4)]

for index, row in df.iterrows():
    code = get_country_code(row['country'])
    if code:
        p = row['percent_deaths']
        cc_population[get_index(p)][code] = p
    else:
        print(f"no code for /{row['country']}/ found")
```

```python
wm = pygal.maps.world.World(height=300)
wm.title = "World Deaths of Confirmed Cases, by Country"
wm.add('.lt.  5%', cc_population[0])
wm.add('.lt. 10%', cc_population[1])
wm.add('.lt. 20%', cc_population[2])
wm.add('.gt. 20%', cc_population[3])
wm.render_to_file('covid.svg')
```
